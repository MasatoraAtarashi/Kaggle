```python
colormap = plt.cm.RdBu
plt.figure(figsize=(12,12))
a = train_df.iloc[:, range(1,30)].join(train_df.SalePrice)
sns.heatmap(a.corr(),linewidths=0.1,vmax=1.0,  square=True, cmap=colormap, linecolor='white', annot=True)
plt.savefig('soukan.png')
```
